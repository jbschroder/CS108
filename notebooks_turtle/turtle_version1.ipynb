### Animation Example

First code block imports lots of stuff for plotting and animation.
Remember, Python is like a blank slate.  You have to import or create what you use.

```python
import matplotlib
%matplotlib inline 

from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.animation
import matplotlib.pyplot as plt
import numpy as np
```

### Define initialize and finalize animation functions

```python
def initialize_animation():
    '''
    Initialize animation environment
    < more documentation>
    '''
    plt.rcParams['figure.dpi'] = 75 
    plt.rcParams["animation.html"] = "jshtml" # javascript html writer
    plt.ioff() # Turn interactive mode off
    plt.rcParams["figure.figsize"] = [5, 5]
    plt.rcParams["figure.autolayout"] = True
    
def finalize_animation():
    '''
    Finalize animation environment
    < more documentation >
    '''
    fig, ax = plt.subplots()
    image = OffsetImage(plt.imread('./turtle.png', format="png"), zoom=0.1)

    def animate(t):
        plt.cla()
        ab = AnnotationBbox(image, movement[t], frameon=False)
        ax.add_artist(ab)
        
        ax.grid()
        plt.xticks(range(10))
        plt.yticks(range(10))
    
    anim = matplotlib.animation.FuncAnimation(fig, animate, frames=len(movement))
    
    return anim
```

### Define library function for creating movements

Could eventually use this block to create a "turtle" object

```python
def move_left():
    last_x, last_y = movement[-1]
    movement.append((last_x-1, last_y))
    
def move_right():
    last_x, last_y = movement[-1]
    movement.append((last_x+1, last_y))

def move_up():
    last_x, last_y = movement[-1]
    movement.append((last_x, last_y+1))
    
def move_down():
    last_x, last_y = movement[-1]
    movement.append((last_x, last_y-1))
```


### This is the code that students would initially write, before 

```python
initialize_animation()

movement = []
movement.append((2,2))

move_left()
move_up()
move_right()
move_down()
    
# Note there is a scope lesson here, that anim must be returned to keep animation alive
anim = finalize_animation()
```
