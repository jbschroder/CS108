```python
# triangle class from earlier lab
class triangle:
        
    # Typically, member variables come first
    side1 = 0
    side2 = 0
    side3 = 0

    # Then, we define member functions
    def __init__(self, s1, s2, s3):
        self.side1 = s1
        self.side2 = s2
        self.side3 = s3
    
    def perimeter(self):
        return self.side1 + self.side2 + self.side3
    
    def area(self):
        # Use heron's formula
        s1_squared = self.side1*self.side1
        s2_squared = self.side2*self.side2
        s3_squared = self.side3*self.side3
        area = 0.25*(4*s1_squared*s2_squared - (s1_squared + s2_squared - s3_squared)**2)**(0.5)
        
        return area
```

```python
t1 = triangle(2,3,4)
print(t1.perimeter(), "perimeter")
print(t1.area(), "area")
```

```python
# now define right triangle
class right_triangle(triangle):    
    # Typically, member variables come first
    
    side1 = 0   # side1 will be the longest
    side2 = 0
    side3 = 0

    # Then, we define member functions
    def __init__(self, s1, s2, s3):
        # fine longest side.
        # 
        if(s1 >= s2 and s1 >= s3):
            self.side1 = s1
            self.side2 = s2
            self.side3 = s3
        elif(s2 >= s1 and s2 >= s3):
            self.side1 = s2
            self.side2 = s1
            self.side3 = s3
        elif(s3 >= s1 and s3 >= s2):
            self.side1 = s3
            self.side2 = s1
            self.side3 = s2
        
        # Because of the if/elif statements above, side1 is always the longest (hypotenuse)
        # Remember, !=   means "not equal"
        if self.side1*self.side1 != (self.side2*self.side2 + self.side3*self.side3):
            print("Error! Sides not correct for a right triangle.")
    
    # a right triangle is an inherited class of triangles, so the area and perimeter functions should still be valid
    # from the parent class.  That is, there is no need to define them differently here
```

```python
t2 = right_triangle(3,4,5)
print(t2.perimeter(), "perimeter")
print(t2.area(), "area")
```

```python
t3 = right_triangle(3,4,4)
print(t3.perimeter(), "perimeter")
print(t3.area(), "area")
```
